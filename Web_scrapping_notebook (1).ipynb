```python
list_col = []
for col in soup.find_all("th"):
    list_col.append(col.text)
df = pd.DataFrame(columns=list_col)# Appending the Data column header of premiership table
list_rs = []
mini_list = []
tr_= soup.find_all("tr")
for i in range(1,len(tr_)):#Looping over the table rows
    mini_list = []# using a mini_list to store each values in table rows
    table_row = tr_[i]
    for row in table_row:
        mini_list.append(row.text)
    list_rs.append(mini_list)
for i in range(len(list_rs)):
    df.loc[len(df)] = list_rs[i]
df.loc[:,"Team name":] 
```

### Checking Data type and converting it to the right suitable

```python
# lets check for the data typpe of the columns
df.dtypes# all the columns are object data type, that make it difficult to perform some operation on them.
#so we need to convert them to the right data type
columns_name = df.columns[2:]
columns_name
```

```python
for col in columns_name:
    df[col] = df[col].astype("int")# to convert fron string to integer
```

### Exploratory Data Analysis

```python
# lets add a new column to the dataframe
```

```python
df["goal_per_match_ratio"] = df["No_of_goals_scored"]/df["matches_played"]
```

```python
df.head()
```

```python
#### Team with the most goals scored per Match, goal_per_match_ratio
## lets get top 5 team with the most goal per match ratio
```

```python
df.groupby("Team_name")["goal_per_match_ratio"].mean().sort_values(ascending = False)[:5].plot(kind = "bar")
plt.title("Team with the most goal_per_match_ratio")
plt.xlabel("Team_name")
plt.ylabel("goal_per_match_ratio");
```

```python
### man city have the highest goal Match_per_goal_ratio, this means that they scored an average of 3 goals per match
```
