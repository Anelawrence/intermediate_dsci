```python
def func_treat(row):
    if "%" in row:
        return float(row.replace("%",''))
    elif "$" in row:
        return int(row.replace("$","").strip().replace(",",""))
    elif "," in row:
        return int(row.replace(",","").replace("$",""))
    return row
```

```python
for columns in columns_to_treat:
    df_2[columns] = df_2[columns].apply(lambda y : func_treat(y))
```
