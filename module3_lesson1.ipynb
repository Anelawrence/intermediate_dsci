```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_alpha_08 (1).csv')
df.head()
```

```python
# Renaming the columns properly
df.columns = [x.replace(' ','_') for x in df.columns]
df.columns
```

```python
df[df.duplicated()]
```

```python
# Using DataFrame.drop_duplicates() to keep first duplicate row
df = df.drop_duplicates(keep='first')
```

```python
df[df.duplicated()]
```

```python
df.info()
```

```python
# data cleaning

df["Cyl"] = df["Cyl"].astype(str).apply(lambda x: x.strip('()').split(' ')[0]).astype(float)

df["Air_Pollution_Score"] = df["Air_Pollution_Score"].astype(str) 
df["Air_Pollution_Score"] = [x[0] for x in df["Air_Pollution_Score"]]  
df["Air_Pollution_Score"] = df["Air_Pollution_Score"].astype(float)    

#df["City MPG"] 
df["City_MPG"] = df["City_MPG"].astype(str).apply(lambda x: x.split('/')[0]).astype(float)

#df["Hwy MPG"] 
df["Hwy_MPG"] = df["Hwy_MPG"].astype(str).apply(lambda x: x.split('/')[0]).astype(float)

#df["Cmb MPG"]
df["Cmb_MPG"] = df["Cmb_MPG"].astype(str).apply(lambda x: x.split('/')[0]).astype(float)

#df["Greenhouse Gas Score"] 
df["Greenhouse_Gas_Score"] = df["Greenhouse_Gas_Score"].astype(str).apply(lambda x: x.split('/')[0]).astype(float)
```

```python
df.isna().sum() # checking for missing values
```

```python
#Filling the null values with either mean or mode

df['Cyl'].fillna(df['Cyl'].mean(), inplace=True)
df['Trans'].fillna(df['Trans'].mode()[0], inplace=True)
df['Drive'].fillna(df['Drive'].mode()[0], inplace=True)
df['FE_Calc_Appr'].fillna(df['FE_Calc_Appr'].mode()[0], inplace=True)
df['City_MPG'].fillna(df['City_MPG'].mean(), inplace=True)
df['Hwy_MPG'].fillna(df['Hwy_MPG'].mean(), inplace=True)
df['Cmb_MPG'].fillna(df['Cmb_MPG'].mean(), inplace=True)
df['Unadj_Cmb_MPG'].fillna(df['Unadj_Cmb_MPG'].mean(), inplace=True)
df['Greenhouse_Gas_Score'].fillna(df['Greenhouse_Gas_Score'].mean(), inplace=True)
```

```python
df.isna().sum() # no more missing values
```

```python
df.to_csv("all_alpha_cleaned")
```

```python
plt.figure(figsize=(10,4))

#Generate a KDE plot for the size feature of the tips dataset. Give detailed explanations of your observations
plt.subplot(1,2,1)
sns.kdeplot(df['Cyl'], shade=True)
plt.title('KDE plot for cyl'.upper())


#Using boxplots generate the distribution of total bill by days considering smoker status of the customer.
plt.subplot(1,2,2)
sns.barplot(data=df, x='Fuel', y='Greenhouse_Gas_Score', ci = 0)
plt.title('barplot of fuel type against Greenhouse Gas Score'.upper())

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```

* Subplot 1: The KDE plot for CYL column is multi-modal and its higest peak between 5 and 7 units.
* Subplot 2: CNG vehicles has the highest greenhouse gas score of over 8 units, while the rest falls between 4 and 6 units.

```python
plt.figure(figsize=(15,6))

#Using the boxplots, generate the distribution of total bill by gender for the tips dataset. Give detailed explanations of your observations
#plt.subplot(1,2,1)
sns.boxplot(data=df, x='Veh_Class', y='Air_Pollution_Score', hue='Fuel')
plt.title('distribution of Air Pollution Score by Vehicle class and fuel'.upper())

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```

Vehicles that uses gasoline has the higest number of air pollution score, followed by ethano/gas vehicles. While vehicles that uses diesel, ethanol and CNG has little or no pollution.
Furthermore, pickup and van has the highest number of air pollution score followed by SUG.

```python
# checking the correlation 
df.corr()
```

```python
#Plot and customize a correlations heatmap for the tips dataset.
corr = df.corr()

# use numpy function to mask the data to create a matrix of booleans
ones = np.ones_like(corr, dtype='bool') # by default returns ones for all the data
# use another numpy function to get just the upper triangle of the boolean array
mask = np.triu(ones)

# adjust mask and correlations 
mask = mask[1:, :-1]
corr = corr.iloc[1:,:-1].copy()

# generate customized heatmap of correlations
fig, ax = plt.subplots(figsize=(10,4))
sns.heatmap(corr, mask=mask, annot=True, fmt=".1f", cmap='Greens', linewidths=.9, cbar_kws={'shrink':.99})
ax.xaxis.tick_bottom()
plt.title("Correlations heatmap of vehicles' data \n".upper(), fontdict={'fontsize': 18}, loc='left')
plt.show()
```

There are strong positive correlations between Greenhouse_Gas_Score, City_MPG, Hwy_MPG, Cmb_MPG and Unadj_Cmb_MPG. Also there is strong negative correlation for (Cyl and Displ) columns against (Greenhouse_Gas_Score, City_MPG, Hwy_MPG, Cmb_MPG and Unadj_Cmb_MPG) columns. While Air Polution score has weak positive correlation across all columns except (Cyl and Displ) columns; which its has weak negative correlation against.
