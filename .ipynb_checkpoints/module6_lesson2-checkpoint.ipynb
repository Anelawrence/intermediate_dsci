```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Nigerian_Car_Prices.csv", index_col=0)
df.head()
```

```python
df.info()
```

```python
df[df.duplicated()] #checking for duplicates. 879 duplicates rows
```

```python
df.drop_duplicates(keep=False, inplace=True) # Drop duplicate rows
df[df.duplicated()]
```

```python
df.isnull().sum() #Checking for null values
```

```python
df['Price'] = df['Price'].astype(object) 
```

```python
df['Price'] = df['Price'].apply(lambda x: x.replace(",","")).astype(float) #cleaning the price column
```

```python
#What is the most expensive Automatic car (show car name and model make) with a mileage of less than 5000

df2 = df[(df['Mileage'] < 5000) & (df['Transmission'] == 'Automatic')][['Make','Price', 'Mileage', 'Transmission']]
df2[df2.Price==df2.Price.max()]
```

```python
#Write a function to bring out the average of each car, which car make has the highest average price? Write a code to show this

def max_avg(col: str):
    df_avgs = df.groupby(col).Price.mean().reset_index()  #grouping the variable column by average price
    return df_avgs[df_avgs.Price == df_avgs.Price.max()]  #return the maximun mean price 
```

```python
max_avg('Make')
```

```python
#Select the row with a price at the 76th percentile score

df[df.Price == np.percentile(df.Price, 76)]  #No price fall on the exact 76th percentile score.
```
